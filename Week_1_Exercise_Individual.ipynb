Dataset

```python
data = [12, 15, 14, 10, 18, 20, 17, 16, 14, 13]
```

1. Print the dataset

```python
print("Original data:", data)
```

2. Compute minimum, maximum, and mean manually

```python
min_value = data[1]
for x in data:
    if x < min_value:
        min_value = x


# Maximum
max_value = data[1]
for x in data:
    if x > max_value:
        max_value = x


# Mean
sum_value = 0
for x in data:
    sum_value += x
    
mean_value = sum_value / len(data)


print("Minimum:", min_value)
print("Maximum:", max_value)
print("Mean:", mean_value)
```

3. Sort the dataset

```python
sorted_data = data
sorted_data.sort()
print(sorted_data)
```

4. Compute the median manually

```python
n = len(sorted_data)
result = n%2
if result == 0:
    median = (sorted_data[n//2] + sorted_data[n//2 - 1]) / 2
else:
    median = sorted_data[n//2 -1]

print("Median:", median)
```

5. Find the most frequent value(s) using a dictionary

```python
frequency = {}
for x in data:
    if x in frequency:
        frequency[x] += 1
    else:
        frequency[x] = 1
print("Frequencies:", frequency)
max_freq = max(frequency.values())
for key, value in frequency.items():
    if value == max_freq:
        most_frequent = key
        break
print("Most Frequent:", most_frequent)
```

TASK 2 – USING PYTHON PACKAGES

```python
import statistics


mean_stat = statistics.mean(data)
median_stat = statistics.median(data)
standard_deviation = statistics.stdev(data)


print("Mean:", mean_stat)
print("Median:", median_stat)
print("Standard Deviation:", standard_deviation)
```

TASK 3 – SIMPLE VISUALIZATION

```python
import matplotlib.pyplot as plt


plt.hist(data, bins=5)
plt.title("Histogram of the Dataset")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

TASK 5 – SMALL FUNCTION (BONUS)

```python
def describe_data(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mean = sum(sorted_data) / n
    
    
    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]


    return {
        "min": min(sorted_data),
        "max": max(sorted_data),
        "mean": mean,
        "median": median
    }


result = describe_data(data)
print(result)
```
